## Matplotlib

[Matplotlib](https://matplotlib.org/) é uma biblioteca para gerar gráficos 2D.  
Foi criada pelo finado [John D. Hunter](https://en.wikipedia.org/wiki/John_D._Hunter) após constatada por este
a ausência de pacotes para plots 2D em Python, sendo sua primeira _release_ de 2003.

````python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

imagem = np.random.rand(10, 10)
print(imagem.shape)
# print(imagem)
plt.imshow(imagem)  # Exibe imagem PIL ou vetor 2D
````

## Organização de um plot
Matplotlib possui a seguinte organização hierárquica:
  - Figure :: ["the whole thing"](https://matplotlib.org/3.1.1/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py)
    - Axes* :: ["what you think of as 'a plot'"](https://matplotlib.org/3.1.1/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py)
      - Axis{2,3} :: ["number-line-like objects"](https://matplotlib.org/3.1.1/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py)
      

Uma _Figure_ é basicamente um container o qual possui zero ou mais instâncias do objeto _Axes_. Este objeto por sua vez é a área na qual será desenhado o gráfico. _Axis_ são as barrinhas com marcações; usualmente, são dois em cada _Axes_ (formando portanto gráficos bidimensionais).

## Criando um plot

Um dos plots mais simples que apresenta alguma informação pode ser criado da seguinte maneira:

````python
x_data = np.array([1, 2, 3, 4])
y_data = np.array([1, 4, 2, 3])
fig, ax = plt.subplots()  # fig: Figure, container
ax.plot(x_data, y_data)  # Plot some data on the axes.
````

Por padrão, ```plt.subplots``` cria uma _Figure_ contendo apenas um _Axes_ ; ou seja, apenas um gráfico.

Os parâmetros nomeados ```nrows``` e ```ncols``` podem ser passados como inteiros positivos caso se deseje mais de um gráfico \( _Axes_ \) na  _Figure_ criada.

````python
fig_mult, ax_mul = plt.subplots(nrows = 2, ncols=3)
print(ax_mul.shape)

# ax_mul é um ndarray pois nrows > 1, e cada elemento é também
# uma lista pois ncols > 1
line1, line2 = ax_mul


for i in range(ax_mul.shape[1]):
    line1[i].plot(np.random.rand(4,1), y_data)
    line2[i].plot(np.random.rand(4,1), y_data)
````

````python
# Criamos uma Figure com 4 axes, 2 linhas e 2 colunas.
fig, axs = plt.subplots(2, 2) # suplots(nrows, ncols) 

# Os Axes criados estão em uma matriz 2x2, então acessamos utilizando índices bidimensionais [i][j]
# Onde i representa a linha e j a coluna, começando em 0, portanto:
axs[0][0].plot([1,2,3,4],[1,2.5,3.6,4])         # Apresenta dados na primeira linha e primeira coluna
axs[0][1].plot([1,2,3,4],[4,3,2,1])             # Apresenta dados na primeira linha e segunda coluna
axs[1][0].plot([1,2,3,3.4,4],[4,2.9,2.1,4,0.3]) # Apresenta dados na segunda linha e primeira coluna
axs[1][1].plot([1,2,3,3.5,4],[1,2.5,2.9,4,2])   # Apresenta dados na segunda linha e segunda coluna
````

### Títulos e Legendas

Ao apresentar um gráfico no matplotlib, é possível configurar diversas propridades, como:

+ Títulos
``` plt.title('título') ```

+ Legendas
``` plt.plot( *args, label='legenda') ```
+ Posição da Legenda
```plt.legend(loc='lower left') ```


> Veja mais atributos para a legenda: https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.legend.html

> Documentação da função plot: https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.plot.html

### Aprensentando Gráficos de Funções 

O matplotlib também pode ser utilizado para apresentar gráficos de funções, utilizando como base os vetores gerados pelo ```linspace``` do numpy. 

Veja o exemplo abaixo, onde serão apresentados os gráficos das funções seno e cosseno.

````python
# Primeiramente, criamos um vetor com valores igualmente espaçados
# entre (start) 0 e (stop) 2*PI, com (num) 360 valores
x = np.linspace(0, 2 * np.pi, 360) # linspace(start, stop, num=50)

# Agora, vamos apresentar a função seno e cosseno.
# Note que passamos como argumento o vetor x, com 360 valores, 
# e as funções serão calculadas para cada um deles.

plt.plot(x, np.sin(x), label='seno')    # apresentando a função seno
plt.plot(x, np.cos(x), label='cosseno') # apresentando a função cosseno

plt.title('Funções')                    # definindo um título para o gráfico
plt.legend(loc='lower left')            # posicionando a legenda no cando inferior esquerdo
````

````python
nrows, ncols = 2, 2
plt.figure(figsize=(ncols * 4, 4))

plt.subplot(nrows, ncols, 1, title='seno')
plt.plot(x, np.sin(x))
plt.subplot(nrows, ncols, 2, title='cosseno')
plt.plot(x, np.cos(x), color='orange')

x10 = np.linspace(0, 2 * np.pi, 10)
plt.subplot(nrows, ncols, 3)
plt.plot(x10, np.sin(x10), 'ro')
plt.subplot(nrows, ncols, 4)
plt.plot(x10, np.cos(x10), 'g+')
````
