### Desenhando uma seta
Pode ser útil para anotar imagens, ou para indicar bem explicitamente alguma relação entre os dados. Ou ainda fazer uma obra de arte, por que não?

````python
ax = plt.axes()
ax.arrow(
    x=0.4, y=0.15,
    dx=0.3, dy=0.25,
    head_width=0.05,
    head_length=0.1,
    facecolor='black',
    edgecolor='blue',
    linewidth='2',
)

plt.show()
````

````python
### Personalizando os gráficos (Axes)
````

````python
x = np.arange(0.0, 2.0, 0.01)
sin = np.sin(2 * np.pi * x)

fig, ax = plt.subplots()

y_right = ax.twinx()  # Cria um eixo y paralelo

ax.plot(x, sin)

ax.tick_params(
    axis='both',  # opção padrão
    labelsize='medium',
    pad=10,  # Espalamento entre o "espinho" preto e o label
)
ax.tick_params(
    axis='x',
    labelcolor='gray',
    labelrotation=30,  # Inclinando em 30º cada label
)

ax.tick_params(
    axis='y',
    labelcolor='#1f8980',
    labelrotation=-30,  # Inclinando em -30º cada label
)
ax.fill(x, sin, facecolor='lightsalmon')
````

### Exibindo imagens
Para carregar uma imagem e exibí-la (e possivelmente anotá-la):

````python
img = mpimg.imread('./img/nature.jpg')
axes = plt.imshow(img)

axes._axes.arrow(x=0, y=0,
    dx=100, dy=200,
    head_width=10,
    head_length=30,
    facecolor='black',
    edgecolor='blue',
    linewidth='2',
)
````

````python
img = mpimg.imread('./img/nature.jpg')
# Aplicando interpolação na imagem
axes = plt.imshow(img, interpolation='bicubic')
````

````python
# Imagem enquanto pseudo-cor
lum_img = img[:, :, 0]
````

Abaixo, a mesma imagem porém sendo aplicado um esquema de pseudo-cores e um mapeamento dessas cores segundo parâmetro "hot" e "nipy_spectral".

````python
plt.imshow(lum_img, cmap='hot')
plt.colorbar()
````

````python
plt.imshow(lum_img, cmap='nipy_spectral')
plt.colorbar()
````
