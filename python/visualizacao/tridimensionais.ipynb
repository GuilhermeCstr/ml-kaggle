## Gráficos Tridimensionais

O matplotlib também pode ser utilizado para apresentar gráficos tridimensionais.

Após incluir o _matplotlib_ e o _numpy_ , devemos também incluir a biblioteca _Axes3D_ de _mplot3d_.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Vejamos, agora, exemplos do uso da biblioteca:

### Gráfico de Linha

São gráficos mais simples, geralmente formados a partir de uma função parametrizada.

```python
# Exemplo extraído de https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html

fig = plt.figure()                               # Cria uma Figure chamada fig
ax = fig.gca(projection='3d')                    # Cria um Axes com projeção 3d.
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)  # Espaço linear entre -4 pi e +4 pi, com 100 divisões
z = np.linspace(-2, 2, 100)                      # Espaço linear que será utilizado no eixo-z, de -2 a +2
r = z**2 + 1                                     # Variável raio: z^2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='curva paramétrica')
ax.legend()                                      # Mostrando a legenda no gráfico

plt.show()
```

### Gráfico de Dispersão

Talvez um dos tipos mais importantes de gráficos a serem utilizados em Aprendizado de Máquina, uma vez que os conjuntos de dados que vão ser utilizados, geralmente, não representam valores contínuos ou descritos por funções matemáticas simples. Portanto, utilizar um gráfico de dispersão facilita a visualizar classificações de dados e suas posições antes e depois do processamento.

Vejamos um exemplo de gráfico de dispersão contendo dois conjuntos de dados com 3 variáveis e duas classificações. Poderiam ser utilizados dados reais, mais utilizamos a distribuição normal da estatística para simular um conjunto com média e desvio padrão bem definidos.

```python
# Exemplo baseado em https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html

fig = plt.figure()                            # Cria uma nova Figure
ax = fig.add_subplot(111, projection='3d')    # Cria um novo Axes com projeção 3D

# Criando um conjunto de dados com uma posição central (x, y, z) e um desvio padrão para cada eixo
# Além disso, utilizaremos um char para representar a cor e outro para o marcador no gráfico.

# Poderíamos ter utilizados dados reais vindos de um processamento de aprendizado de máquina.

# data= [xcenter, ycenter, zcenter, xdev, ydev, zdev, color, marker]
dataA = (24,   0, -30, 3, 33, 8, 'r', 'o')          # o marcador 'o' indica circulos no gráfico; 'r' = cor vermelha
dataB = (32, -40, -40, 3, 33, 8, 'b', '^')          # o marcador 'b' indica triângulos no gráfico; 'b' = cor azul 

# Número de pontos para cada conjunto de dado
n = 58

for xcenter, ycenter, zcenter, xdev, ydev, zdev, color, marker in [dataA, dataB]:
    xs = np.random.normal(xcenter, xdev, n)         # a função normal da biblioteca random do numpy
    ys = np.random.normal(ycenter, ydev, n)         # gera valores com uma distribuição normal
    zs = np.random.normal(zcenter, zdev, n)         # : normal(mean, sigma, count)
    ax.scatter(xs, ys, zs, c=color, marker=marker)

# Definindo labels dos eixos x, y e z.
ax.set_xlabel('eixo X')
ax.set_ylabel('eixo Y')
ax.set_zlabel('eixo Z')

# Apresentando gráfico.
plt.show()
```
